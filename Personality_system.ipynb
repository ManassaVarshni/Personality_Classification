### EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="dark")

# Plotting the top 10 countries with highest number of participants

demographic = pd.DataFrame(data['country'].value_counts())
demographic.reset_index(level = 0, inplace = True)
demographic.columns = ['Country', 'Number of participants']
max_no_of_participants = demographic[0:10]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
sns.barplot(x = max_no_of_participants['Country'], y = max_no_of_participants['Number of participants'], ax = ax, edgecolor = 'black')

plt.title("COUNTRIES WITH HIGHEST NUMBER OF PARTICIPANTS")
plt.xlabel("Countries")
plt.ylabel("Number of participants")
plt.show()
```

```python
# Function to visualize all the 10 questions in each five traits

def visualize_questions(start_col, end_col, color, dic):
    n = 1
    plt.figure(figsize=(15,30))
    for col in data.iloc[:, start_col:end_col].columns:
        plt.subplot(5, 2, n)
        sns.countplot(x = col, data = data, color = color, edgecolor = 'black')
        plt.title(col + " : " + dic[col])
        n += 1
    plt.show()
    
```

### KMEANS CLUSTERING

Here, we will be dividing the data into 10 personality types(we can choose a number of our own) and calculate the scores of the five traits in each type. To do this, we will be using minibatchkmeans.

```python
data_model = pd.DataFrame(data.iloc[:, :50])
# data_model = data_model.fillna(0)
data_model.head()
```

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
kmeans = MiniBatchKMeans(n_clusters = 10, random_state = 0, batch_size = 100, max_iter = 150).fit(data_model)
```
